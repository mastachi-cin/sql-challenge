```python
# Dependencies
import os
from sqlalchemy import create_engine
import matplotlib.pyplot as plt
from dotenv import load_dotenv
import numpy as np
```

```python
# Get user and password from .env file
load_dotenv()
usern = os.environ.get('USERNAME')
passw = os.environ.get('PASSWORD')

# Database info
database_url = f'postgresql://{usern}:{passw}@localhost:5432/employees_db'
    
# Database Connection
engine = create_engine(database_url)
connection = engine.connect()

# Salary query
result = connection.execute("select * from salaries")
```

```python
# List for salaries
salaries = []
for row in result:
    salaries.append(row['salary'])
    
result.close()

# Histogram for visualizing salary ranges
plt.hist(salaries)
plt.title('Salaries Distribution')
plt.xlabel("Salary")
plt.ylabel('Count')
plt.show()
```

```python
# Average salary by title query
query = """
        select t.title_name, round(avg(salary),2) as average_salary
        from salaries s
        inner join employees e on e.emp_no = s.emp_no
        inner join titles t on t.title_id = e.title_id
        group by title_name
        """

# Query execution
ave_salary = connection.execute(query)

# Lists for average salary and title name
title_name_ls, ave_salary_ls = zip(*[(row['title_name'], float(row['average_salary'])) for row in ave_salary])

ave_salary.close()

# Bar chart of average salary by title

# List for x axis
x_axis = np.arange(len(title_name_ls))

# Bar chart
plt.bar(x_axis, ave_salary_ls, align="center", label="Average salary")

# Place x axis headers
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, title_name_ls, rotation="vertical")

# Set legend
plt.legend(loc="best")

# Give the chart a title, x label, and y label
plt.title("Average salary by Title")
plt.xlabel("Title name")
plt.ylabel("Average salary")

# Print chart to screen
plt.show()
```
